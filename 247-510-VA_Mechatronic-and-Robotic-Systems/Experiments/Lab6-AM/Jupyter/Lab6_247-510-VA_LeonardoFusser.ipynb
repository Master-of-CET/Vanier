The purpose of this work is to demonstrate how an AM signal can be generated in simulation and show the effect Gaussian noise has on an AM signal. We will use many parameters for this simulation, they are shown below.

```python
import numpy as np
import matplotlib.pyplot as plt

fc = 1E3
fs = 1E2
Vc = 1
Vn = 0.2
Vs = 0.5
N = 500
t = np.linspace(0, 10E-3, N)
```

We will use *N* points to perform our simulation.

```python
carrier = Vc*np.sin(2*np.pi*fc*t)
signal = Vs*np.sin(2*np.pi*fs*t)
noise = Vn*np.random.randn(N)
AM_SIGNAL = signal*carrier
AM_SIGNAL_N = signal*carrier+noise
```

First, we will look at the noise and carrier signals by superimposing them on the same plot.

```python
plt.plot(t, carrier, '--', label='Carrier')
plt.plot(t, noise, label='Noise')
plt.legend()
plt.title('Carrier Signal (%6.2fV, f=%6.2fHz )'%(Vc, fc))
plt.xlabel('Time (Seconds)')
plt.ylabel('Amplitude (Volts)')
```

The noise is *Gaussian*. We know this because of the histogram it produces. It peaked near 0 and tapers off at higher and lower voltages.

```python
plt.hist(noise, bins=20)
plt.title('Noise Distribution')
```

The AM signal itself can be plotted. We will show the carrier with the signal as its' envlope with the dotted line.

```python
plt.plot(t, AM_SIGNAL, 'g', label='Signal and Carrier')
plt.plot(t, signal, 'r:', label='Signal Envelope')
plt.legend()
plt.plot(t, -signal, 'r:')
plt.title('AM Signal')
plt.xlabel('Time (Seconds)')
plt.ylabel('Amplitude (Volts)')
```
